Importación de las librerías:

```python
import pandas as pd
import numpy as np
from plotnine import *
```

Lectura de las bases de datos y unión de estás a través de la columna id:

```python
d1 = pd.read_csv("nene1.txt", sep = ";")
d2 = pd.read_csv("nene2.txt", sep = ";")
df = pd.merge(d1, d2, on = "id", how = "inner")
df.info()
```

Filtro para selección solo la región de Aysén:

```python
#Como se indica en el informe entregado, la región de Aysén corresponde a la número 11 de la base de datos entregada.
df_aysen = df[df["region"] == "11"] 
df_aysen
```

Revisión de los valores para las columnas:

```python
df_aysen["sexo"].unique()
```

```python
#Al tener distintos valores para representar lo mismo como hombre con 1 y mujer con 2, esto puede generar
#dificultad al momento de manipular la base de datos, por lo que es mejor usar un único sistema, en nuestro caso
#dejar Hombre y Mujer.
df_aysen["sexo"].replace([1, 2], ["Hombre", "Mujer"], inplace = True)
df_aysen["sexo"].unique()
```

Revisión de la variable "b14_rev4cl_caenes":

```python
df_aysen["b14_rev4cl_caenes"].unique()
```

```python
#Al encontrar un valor no registrado en la base de datos lo analizamos.
df_aysen[df_aysen["b14_rev4cl_caenes"] == 999]
```

Podemos observar el valor 999, el cual no aparece registrado en el informe entregado. Lo asumiremos como rama de actividad económica en categoría de "otros".

Revisión de las columnas y su cantidad de valores nulos, duplicados:

```python
print("Cantidad de duplicados: ", df_aysen.duplicated().sum())
df_aysen.isnull().sum()
```

No se encontraron datos duplicados, pero si valores nulos, los que procederemos a analizar.

```python
#Creamos una nueva base de datos llamada "reemplazados" igual a la base de datos "df_aysen"
#y le aplicamos un filtro para rellenar todos los valores nulos con un "9", de esta forma
#los seleccionamos y analizamos las filas.
reemplazados = df_aysen
reemplazados["ocup_form"].fillna(9, inplace = True)
reemplazados[reemplazados["ocup_form"] == 9]
```

Al no encontrar un valor particular, podemos ver que hay personas menores a 15 años los cuales tienen este campo en nulo, lo cual tiene sentido.
